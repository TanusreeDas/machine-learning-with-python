```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 9, 10]
plt.figure(figsize=(10,10))
plt.plot(x,y,'ro--', label='Line A')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot')
plt.legend()
plt.grid(True)
plt.show()
```
